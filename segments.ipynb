```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
outlet_stats_df = pd.read_csv('outlet_stats.csv')
encoded_outlets_df = pd.read_csv('encoded_outlets.csv')

# Merge the two dataframes
outlet_stats_df = outlet_stats_df.merge(encoded_outlets_df, on='Outlet_ID')
outlet_stats_df = outlet_stats_df.drop(columns=['0', '1', '2'])
outlet_stats_df.sort_values(by=['cluster'], inplace=True)
outlet_stats_df.head()
```

```python
# plot area by outlet for each cluster
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['area(sqft)'], c=outlet_stats_df['cluster'])
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [18, 20]

plt.subplot(3, 3, 1)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['area(sqft)'], c=outlet_stats_df['cluster'])
plt.title('Area by Outlet')

plt.subplot(3, 3, 2)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['units_totals'], c=outlet_stats_df['cluster'])
plt.title('Total Units sold by Outlet')

plt.subplot(3, 3, 3)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['volume_totals'], c=outlet_stats_df['cluster'])
plt.title('Total Volume sold by Outlet')

plt.subplot(3, 3, 4)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['volume_week7'], c=outlet_stats_df['cluster'])
plt.title('Volume sold in Week 7 by Outlet')

plt.subplot(3, 3, 5)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['volume_IP1'], c=outlet_stats_df['cluster'])
plt.title('Volume sold in IP1 by Outlet')

plt.subplot(3, 3, 6)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['volume_BP2'], c=outlet_stats_df['cluster'])
plt.title('Volume sold in BP2 by Outlet')

plt.subplot(3, 3, 7)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['units_BP4'], c=outlet_stats_df['cluster'])
plt.title('Units sold in BP4 by Outlet')

plt.subplot(3, 3, 8)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['units_IP3'], c=outlet_stats_df['cluster'])
plt.title('Units sold in IP3 by Outlet')

plt.subplot(3, 3, 9)
plt.scatter(outlet_stats_df['Outlet_ID'], outlet_stats_df['revenue_month2'], c=outlet_stats_df['cluster'])
plt.title('Revenue in Month 2 by Outlet')

plt.show()
```
